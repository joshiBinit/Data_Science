```python
df3['tit']=df3.Title.astype('str').apply(lambda x: x[:7])
```

```python
df3.head()
```

```python
del df3['Road']
del df3['Views']
```

```python
def parse_road_width(x):
    qty,measure=x.split(' ')[0],x.split(' ')[1]
    if measure =='Feet':
        return float(qty)*0.3048
    return float(qty)
df3['road_width_meters']=df3['Road Width'].astype('str').apply(parse_road_width)
del df3['Road Width']
```

```python
df3['adress0']=df3.Address.astype('str').apply(lambda x: x.split(',')[0])
df3['adress1']=df3.Address.astype('str').apply(lambda x: x.split(',')[1])
df3['adress2']=df3.Address.astype('str').apply(lambda x: x.split(',')[2])
del df3['Address']
print(len(df3.adress0.unique()),len(df3.adress1.unique()),len(df3.adress2.unique()),len(df3.City.unique()))
```
